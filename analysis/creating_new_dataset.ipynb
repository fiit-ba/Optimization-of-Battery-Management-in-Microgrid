```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/data_processed_cleaned.csv")
df['date'] = pd.to_datetime(df['date'], format='%Y/%m/%d %H:%M')

df['fve'] = df['fve'] * 0.03
df['mve'] = df['mve'] * 0.07
df['wpg'] = df['wpg'] * 0.07

df['production_usage'] = df['fve'] + df['mve'] + df['wpg'] 
```

```python
df.head()
```

```python
columns_to_modify = ['shops', 'office_building', 'industrial_park', 'fve', 'mve','wpg', 'price','temp', 'wind']
not_negative = ['shops', 'office_building', 'industrial_park', 'fve', 'mve','wpg']
mod_df = df.copy()
for c in columns_to_modify:
    
    #mu = mean
    # sigma = standard deviation
    mu = 0
    sigma = 10
    noise = np.random.normal(mu, sigma, [df[c].shape[0]]) 
    mod_df[c] = df[c] + noise

for c in not_negative:
    mod_df[c] = mod_df[c].apply(lambda x : 0 if x < 0 else x)
```
